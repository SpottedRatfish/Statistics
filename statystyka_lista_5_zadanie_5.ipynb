```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3
n = np.arange(50,2000,100)
M = 1000
```

```python
def samples(size, a):
    samplesl = []
    for i in range(0, size):
        u = np.random.uniform()
        samplesl.append(u**(1/(a+1)))
    return samplesl
```

```python
def estymacje(n, M):
    MNWl = []
    MMl = []
    
    for i in n:
        MNW = []
        MM = []
        for j in range(M): 
            u = np.random.random(i)
            sample = u**(1/(a+1))
            MNW.append(len(sample)/(-np.sum(np.log(sample)))-1)
            MM.append((2*np.mean(sample) - 1)/(1 - np.mean(sample)))
        MNWl.append(MNW)
        MMl.append(MM)
    return MNWl, MMl

MNW_estymator, MM_estymator = estymacje(n, M)
```

```python
MNW_med = []
MM_med = []
MNW_var = []
MM_var = []

for i in range(len(MNW_estymator)):
    MNW_med.append(np.median(MNW_estymator[i]))
    MNW_var.append(np.var(MNW_estymator[i])) 
for i in range(len(MM_estymator)):
    MM_med.append(np.median(MM_estymator[i]))
    MM_var.append(np.var(MM_estymator[i])) 
```

```python
    
plt.plot(n,MNW_med, label="MNW")
plt.plot(n,MM_med, label="MM")
plt.hlines(a, xmin = n[0], xmax = n[-1] ,linestyle = 'dashed', color = 'black', label = 'mediana' )
plt.legend()
plt.show()
```
