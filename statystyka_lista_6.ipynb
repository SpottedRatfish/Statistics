```python
import numpy as np
from scipy.stats import norm, t, chi2
```

```python
n1= 20
n2 = 50
n3 = 100
normal_1 = np.random.normal(2.1,0.2,n1)
normal_2 = np.random.normal(2.1,0.2,n2)
normal_3 = np.random.normal(2.1,0.2,n3)
mi = 2.1
sigma = 0.2
```

```python
def check(liczba):
    n = 1000
    count = 0
    mi = 2.1
    sigma = 0.2
    for i in range(n):
        normal_emp = np.random.normal(mi, sigma, liczba)
        brzeg_lewy = np.mean(normal_emp) - (sigma/np.sqrt(liczba))*norm.ppf(0.99)
        brzeg_prawy = np.mean(normal_emp) + (sigma/np.sqrt(liczba))*norm.ppf(0.99)
        if brzeg_lewy < mi < brzeg_prawy :
            count += 1
    return count
```

```python
check(n1)
```

```python
check(n2)
```

```python
check(10000)
```

```python
def check_2(liczba):
    n = 1000
    count = 0
    mi = 2.1
    sigma = 0.2
    for i in range(n):
        normal = np.random.normal(mi, sigma, liczba)
        brzeg_lewy = np.mean(normal) - (np.std(normal)/np.sqrt(liczba))*t.ppf(0.99,11)
        brzeg_prawy = np.mean(normal) + (np.std(normal)/np.sqrt(liczba))*t.ppf(0.99,11)
        if brzeg_lewy < mi:
            if brzeg_prawy > mi:
                count += 1
    return count
```

```python
check_2(n1)
```

```python
check_2(n2)
```

```python
check_2(n3)
```
