```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
```

```python
count = 0
for i in range(1000):
    x = np.random.standard_t(1,1000)
    count += stats.shapiro(x)[1]
```

```python
count_final = []
for j in range(100):
    count = 0
    for i in range(1000):
        x = np.random.standard_t(j+1,1000)
        count += stats.shapiro(x)[1]
    count_final.append(count)
```

```python
count_final = np.array(count_final) / 1000
count_final = count_final*(-1)
count_final+= 1
```

```python
count_final2 = []
for j in range(100):
    count2 = 0
    for i in range(1000):
        x = np.random.standard_t(j+1,1000)
        count2 += stats.kstest(x,stats.norm.cdf)[1]
    count_final2.append(count2)

```

```python
count_final2 = np.array(count_final2) / 1000
count_final2 = count_final2*(-1)
count_final2 += 1
```

```python
plt.figure(figsize = (14,8))
plt.xlabel("wartoci parametrw dla rozkadu t-studenta")
plt.plot(count_final, label = "Shapiro-Wilk")
plt.plot(count_final2, label = "Komogorow")
plt.legend()
```
