```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, lognorm, pareto, expon
S_all = []
for i in range(1000):
    n= 100
    mi = 1
    sigma = 0.1
    M = np.random.lognormal(mi,sigma,n)
    S_all.append(np.sum(M)/n)
```

```python
S_all_2 = []
for i in range(1000):
    n= 500
    mi = 1
    sigma = 0.1
    M = np.random.lognormal(mi,sigma,n)
    S_all_2.append(np.sum(M)/n)
```

```python
data = pd.DataFrame({"n=100":S_all,"n=500" : S_all_2})
ax = data[['n=100','n=500']].plot(kind = 'box', title = 'boxplot')
x = np.linspace(0,500,1000)
ax.axhline(np.mean(np.random.lognormal(mi,sigma,10000)))
```

```python
#zadanie 2
mi = 0
sigma = 1
n = 1000
max_normal = []
M = 1000
for i in range(M):
    X_normal = np.random.normal(mi,sigma,n)
    max_normal.append(max(X_normal))
max_normal = np.array(max_normal)    
```

```python
cdfx = np.sort(max_normal)
x_val = np.linspace(start=min(cdfx),stop=max(cdfx),num=len(cdfx))
size = max_normal.size
y_val = []
for i in x_val:
    temp = max_normal[max_normal <= i]
    value = temp.size / size
    y_val.append(value)

```

```python
plt.plot(x_val,np.array(y_val), label = "dystrybuanta empiryczna", color = "blue")
plt.plot(x_val, (norm.cdf(x_val))**n, color = 'red', label = "dystrybuanta teoretyczna")
plt.legend()
```

```python
n = 1000
M = 1000
pareto_max = []
for i in range(M):
    U = np.random.random(1000)
    U = 5*(U**(-1/5)-1)
    pareto_max.append(max(U))
pareto_max = np.array(pareto_max)
```

```python
def dist_teo_pareto(lam,alph,x_val):
    y = 1 - (lam/(lam+x_val))**alph
    return y


cdfx = np.sort(pareto_max)
x_val = np.linspace(start=min(cdfx),stop=max(cdfx),num=len(cdfx))
size = pareto_max.size
y_val = []
for i in x_val:
    temp = pareto_max[pareto_max <= i]
    value = temp.size / size
    y_val.append(value)
```

```python
plt.plot(x_val,(1 - (5/(5+x_val))**5)**n,label = "dystrybuanta teoretyczna")
plt.plot(x_val,np.array(y_val), label = "dystrybuanta empiryczna")
plt.legend()
```

```python
#zadanie 10
mi = 0
sigma = 1
theta_1 = []
theta_2 = []
for i in range(1000):
    X = np.random.normal(mi,sigma,7)
    theta_1.append(np.sum(X)/7)
    theta_2.append((2*X[0]-X[5]+X[3])/2)
```

```python
theta_1 = np.array(theta_1)
theta_2 = np.array(theta_2)
```

```python
cdfx = np.sort(theta_1)
x_val = np.linspace(start=min(cdfx),stop=max(cdfx),num=len(cdfx))
size = theta_1.size
y_val = []
for i in x_val:
    temp = theta_1[theta_1 <= i]
    value = temp.size / size
    y_val.append(value)
```

```python
cdfx = np.sort(theta_2)
x_val_2 = np.linspace(start=min(cdfx),stop=max(cdfx),num=len(cdfx))
size = theta_2.size
y_val_2 = []
for i in x_val:
    temp = theta_2[theta_2 <= i]
    value = temp.size / size
    y_val_2.append(value)
```

```python
plt.plot(x_val, np.array(y_val),label = "dystrybuanta - theta1")
plt.plot(np.linspace(-4,4,1000),norm.cdf(np.linspace(-4,4,1000),mi,1/7*sigma),label = "dystrybuanta teoretyczna")
plt.legend()
```
